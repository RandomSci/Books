that question? <br>

• What features of the data did I extract, and will these enable the right

• How will the machine learning solution interact with other parts of my research
or business product? </h4>

# A First Application: Classifying Iris Species



```python
import matplotlib.pyplot as plt, numpy as np, pandas as pd
from pandas.plotting import scatter_matrix
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import mglearn

iris_dataset = load_iris()

X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=0)

df = pd.DataFrame(X_train, columns=iris_dataset['feature_names'])
```

```python
pair_plots = scatter_matrix(df, c = y_train, marker='0', figsize=(10, 10), hist_kwds={"bins": 20}, s = 60, alpha=.8, cmap=mglearn.cm3)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()

knn.fit(X_train, y_train)

new = np.array([[5, 2.9, 1, 0.2]])

predict = knn.predict(new)
predicted_object = iris_dataset['target_names'][predict][0]
print(predicted_object)
```

```python
predict_test = knn.predict(X_test)
test_set_score = knn.score(X_test, y_test)
print(f"Test set score: {test_set_score * 100:.2f}%")
```
