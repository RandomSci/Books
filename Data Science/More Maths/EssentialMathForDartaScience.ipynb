<h2>Euler’s Number and Natural Logarithms</h2>

<p>here is a special number that shows up quite a bit in math called Euler’s <br>
number <i>e</i> . It is a special number much like Pi π and is approximately <br>
2.71828. <i>e</i> is used a lot because it mathematically simplifies a lot of <br>
problems. We will cover e in the context of exponents and logarithms.</p>

```python
import numpy as np; np.e
```

```python
from sympy import symbols, Eq, solve

A, P, r, n, t = symbols("A P r n t")

formula = Eq(A, P * (1 + r / n)**(n * t))

values = {P: 100, r: 0.20, n: 525600 , t: 2}

result = solve(formula.subs(values), A)

display(formula)
result[0]
```

<h2>Natural Logarithms</h2>
<p>
When we use e as our base for a logarithm, we call it a natural logarithm. <br>
Depending on the platform, we may use <b><i>ln()</i></b> instead of <b><i>log()</i></b> to specify <br>
a <b>natural logarithm</b>. So rather than express a natural logarithm expressed as <br> 
log<sub>e</sub>10 to find the power raised on e to get 10, we would shorthand it as <br>
<i>ln</i>(10):
</p>

```python
from sympy import *

e, log, ln = symbols('e log ln')

eq = f"{log}{e}10 = {ln}(10)"
print(eq)
```

<p>
However, in Python, a natural logarithm is specified by the <b><i>log()</b></i> <br>
function. As discussed earlier, the default base for the <b><i>log()</b></i> function is <b><i>e</b></i>. <br>
Just leave the second argument for the base empty and it will default to <b>
using <b><i>e</b></i>
</p>

```python
from math import log 
 
# e raised to what power gives us 10?
x = log(10) 
 
print(x)
```

<h2>Limits</h2>



```python
# Using SymPy to calculate limits
from sympy import *

x = symbols("x")
f = 1 * x
result = limit(f, x, 00) # (00 means infinity in python library sympy)
print(result)
```

```python
from sympy import * 
 
n = symbols('n')
f = (1 + (1/n))**n
result = limit(f, n, oo) 

print(result) # E
print(result.evalf())
```

# Derivatives

<p>
A derivative tells the slope of a <br>
function, and it is useful to measure the rate of change at any point in a <br>
function. <br>
Why do we care about derivatives? They are often used in machine learning <br>
and other mathematical algorithms, especially with gradient descent. When <br>
the slope is 0, that means we are at the minimum or maximum of an output <br>
variable.
</p>

```python
def derivative_x(f, x, step_size): 
    m = (f(x + step_size) - f(x)) / ((x + step_size) - x) 
    return m 
 
 
def my_function(x): 
    return x**2 
 
slope_at_2 = derivative_x(my_function, 2, .00001) 
 
print(slope_at_2)  # page 54
```
